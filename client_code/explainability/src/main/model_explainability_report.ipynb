# model_explainability_report.ipynb
----------------------------------
Purpose:
    Notebook to generate visual and textual explainability reports
    for AI/ML models used in FundWise.
Features:
- Compute SHAP values for all users/funds
- Generate feature importance plots
- Export readable explanations for dashboard integration
- Highlight anomalies detected by graph-based methods
- Compare contributions, scores, and risk tags

Example Workflow:
1. Load trained credibility scoring model
2. Load transaction, behavior, and graph features
3. Run SHAPExplainer for all users
4. Generate ScoreReasonGenerator explanations
5. Aggregate reports for moderators and UI dashboards
